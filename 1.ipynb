   - **Непрерывные переменные:** могут принимать любое значение в диапазоне.
   - **Дискретные переменные:** принимают только определенные значения (обычно целые числа).

   **Ваши ответы:**

   - **Непрерывные переменные:**
     - `[age,
        resting blood pressure,
        serum cholestoral in mg/dl,
        maximum heart rate achieved,
        oldpeak,
        number of major vessels
      ]` ⏰
   
   - **Дискретные переменные:**
     - `[sex,
        chest paint type,
        fasting blood sugar > 120 mg/dl,
        resting electrocardiographic results
        exercise induced angina,
        slope of peak,
        thal,
        disease
        ]` ⏰

Коментарии
1. Chest pain type - вероятнее всего представляет собой категории/типы болей в груди, в этом случае - дискретная. Так же может обозначать интенсивность боли, в этом случае велечина непрерывная
2. Resting electrocardiographic results - типы результатов ЭКГ в состоянии покоя, скорее всего дискретная
3. slope of peak - типы наклона сегмента ST, всего три уникальных значения обозначающие тип наклона
4. oldpeak - депрессия сегмента ST, 
5. thal - Таллиевый стресс-тест, помогает оценить наличие ишемии миокарда, три уникальных значения: 3 - нормальный, 6 - исправленный дефект, 7 - сильный дефект
6. number of major vessels - количество сосудов, может быть как и непрерывной так и дискретной

## **4. Визуализация распределения переменных**


1. Гистограмма для числовой переменной

```python
sns.histplot(df['age'], kde=True)
plt.title('Распределение age')
plt.xlabel('Age')
plt.ylabel('Частота')
plt.show()
```

2. Столбчатая диаграмма для категориальный переменной

```python
sns.countplot(x='chest pain type', data=df)
plt.title('Распределение chest pain type')
plt.xlabel('chest pain type')
plt.ylabel('Количество')
plt.show()
```
