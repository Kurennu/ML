   - **Непрерывные переменные:** могут принимать любое значение в диапазоне.
   - **Дискретные переменные:** принимают только определенные значения (обычно целые числа).

   **Ваши ответы:**

   - **Непрерывные переменные:**
     - `[age,
        resting blood pressure,

      ]` ⏰
   
   - **Дискретные переменные:**
     - `[sex,
        chest paint type,
        fasting blood sugar > 120 mg/dl,
        resting electrocardiographic results
        exercise induced angina,
        slope of peak,
        thal,

        ]` ⏰

Коментарии
1. Chest pain type - вероятнее всего представляет собой категории/типы болей в груди, в этом случае - дискретная. Так же может обозначать интенсивность боли, в этом случае велечина непрерывная
2. Resting electrocardiographic results - типы результатов ЭКГ в состоянии покоя, скорее всего дискретная
3. slope of peak - типы наклона сегмента ST, всего три уникальных значения обозначающие тип наклона
4. oldpeak - депрессия сегмента ST, 
5. thal - Таллиевый стресс-тест, помогает оценить наличие ишемии миокарда, три уникальных значения: 3 - норма или незначительный дефект, 6 - исправленный дефект, 7 - сильный дефект
6. number of major vessels - количество сосудов, может быть как и непрерывной так и дискретной

### **Визуализация распределения переменных**


1. Гистограмма для числовой переменной Age

```python
sns.histplot(df['age'], kde=True)
plt.title('Распределение age')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()
```

Большая часть людей в выборке в возрасте от 40 до 65, что в целом типичная возрастная категория для пациентов с сердечными заболеваниями. Так же аблица не содержит информации о молодых людях до 30 лет

2. Столбчатая диаграмма для категориальный переменной thal

```python
sns.countplot(x='thal', data=df)
plt.title('Распределение thal')
plt.xlabel('Таллиевый стресс-тест')
plt.ylabel('Количество')
plt.show()
```

Наиболее часто встречается результат 3, это считается нормой или незначительным дефектом. Это означает, что у большинства пациентов есть признаки излечимой ишемии миокарда. Категория 7 - необратимый дефект, встречается вторым по частоте, возможно эта часть пациентов перенесла инфаркт

3. Боксплот для числовой переменной serum cholestoral in mg/dl:

```python
sns.boxplot(x=df['serum cholestoral in mg/dl      '])
plt.title('Boxplot serum cholestoral in mg/dl')
plt.xlabel('serum cholestoral in mg/dl')
plt.show()
```

Нету отрицательных и черезмерно высоких значений, хороший сбор данных, но есть пару выбросов в высоких значениях. Нормальный уровень холестерина обычно считается ниже 200 мг/дл, значительная часть выборки имеет повышенный уровень холестерина, что указывает на повышенный риск сердечно-сосудистых заболеваний.
